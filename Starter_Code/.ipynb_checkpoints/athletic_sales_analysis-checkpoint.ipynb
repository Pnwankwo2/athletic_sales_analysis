```python
# Resample the pivot table into daily bins, and get the total sales for each day.
#combined_df = combined_df.sort_values(by='invoice_date')
combined_df = combined_df.set_index('invoice_date')
invoice_sales_pivot = combined_df.pivot_table(
    values='total_sales', 
    index='invoice_date', 
    aggfunc='sum'
).reset_index()

daily_sales = invoice_sales_pivot.resample('D').sum().reset_index()


# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales = daily_sales.sort_values(by='total_sales', ascending=False)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Ensure invoice_sales_pivot is defined
# Ensure 'total_sales' column exists
if 'total_sales' not in combined_df.columns:
    combined_df['total_sales'] = combined_df['Total_Products_Sold'] * combined_df['price']

invoice_sales_pivot = combined_df.pivot_table(
    values='total_sales', 
    index='invoice_date', 
    aggfunc='sum'
).reset_index()

# Resample the pivot table into weekly bins, and get the total sales for each week.
invoice_sales_pivot['invoice_date'] = pd.to_datetime(invoice_sales_pivot['invoice_date'])

# Set 'invoice_date' as the index
invoice_sales_pivot.set_index('invoice_date', inplace=True)

# Resample into weekly bins and get the total sales for each week
weekly_sales = invoice_sales_pivot.resample('W').sum().reset_index()

# Display the weekly sales
print(weekly_sales.head())

# Sort the resampled pivot table in descending order on "Total Sales".
```
